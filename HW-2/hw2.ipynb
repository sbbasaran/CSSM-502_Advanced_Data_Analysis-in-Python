```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import statsmodels.api as sm
from matplotlib import pyplot as plt

import os
```

```python
data = pd.read_csv("/Users/ideakadikoy/Desktop/veri.csv")
data
```

```python
data = data[['E3009', 'E2013']].dropna()

data["E3009"] = pd.Categorical(data["E3009"]).codes
data["E2013"] = pd.Categorical(data["E2013"]).codes

y = data['E2013']
x = pd.DataFrame(data['E3009'])


regres1 = sm.OLS(y, x).fit()
```

```python
print(regres1.summary().as_latex())

fig = sns.regplot(x="E3009", y="E2013", data=data)
sns.set_style("whitegrid")
plt.title("Regression Results")
plt.xlabel('GOVERNMENT PERFORMANCE: GENERAL')
plt.ylabel('RELIGIOUS DENOMINATION')
plt.show()
```
